# Make TSNE embeddings CSV for Qian

```python
import pandas as pd
import numpy as np

# Load contextualized TSNE embeddings
char_tsne_fpath = '/usr0/home/mamille2/erebor/fanfiction-project/output/tsne_chars_lc10_context.nptxt'
char_labels_fpath = '/usr0/home/mamille2/erebor/fanfiction-project/output/tsne_chars_lc10_context_labels.nptxt'
char_fnames_fpath = '/usr0/home/mamille2/erebor/fanfiction-project/output/tsne_fnames_lc10_context.nptxt'

tsne_embs = np.loadtxt(char_tsne_fpath)
tsne_embs.shape
```

```python
with open(char_labels_fpath, 'r') as f:
    char_labels = f.read().splitlines()
    print(len(char_labels))
```

```python
with open(char_fnames_fpath, 'r') as f:
    char_fnames = f.read().splitlines()
    print(len(char_fnames))
```

```python
outlines = [[tsne_x, tsne_y, char, fname] for tsne_x, tsne_y, char, fname in zip(tsne_embs[:,0], tsne_embs[:,1], char_labels, char_fnames)]
out_df = pd.DataFrame(outlines, columns=['tsne_x', 'tsne_y', 'character', 'filename'])
out_df
```

```python
out_df.to_csv('/usr0/home/mamille2/erebor/fanfiction-project/output/tsne_lc10_context.csv', index=False)
```

# Check overlap between Qian's requested fics and dataset_1k-5k

```python
import pandas as pd

# Load Qian's requested fics
fics_fpath = '/usr0/home/mamille2/erebor/fanfiction-project/data/stories_authors_sample1_300.csv'
fics = pd.read_csv(fics_fpath, header=None, index_col=0, names=['fic_id'])
print(fics.columns)
print(len(fics))
```

```python
fic_id_outpath = '/usr0/home/mamille2/erebor/fanfiction-project/data/qian_sample1_fic_ids.csv'
fics.to_csv(fic_id_outpath, index=False)
```

```python
# Load dataset_1k-5k fics
metadata = pd.read_csv('/usr0/home/mamille2/erebor/fanfiction-project/data/ao3/harrypotter/dataset_1k-5k/metadata.csv')
metadata.columns
```

```python
dataset_fics = metadata['fic_id'].tolist()
qian_fics = fics['fic_id'].tolist()
print(len(dataset_fics))
print(len(qian_fics))
print(len(set(dataset_fics).intersection(set(qian_fics))))
```
