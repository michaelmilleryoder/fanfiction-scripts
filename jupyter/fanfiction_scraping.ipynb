# Exclude fic_ids that were already scraped

## Already scraped is in AO3Scraper/output

```python
# Load new stories
import os

dirpath = '/projects/AO3Scraper/output/ao3_dcu_text/stories'
story_ids = [int(fname[:-4]) for fname in os.listdir(dirpath)]
len(story_ids)
```

```python
def remove_bytestring(value):
    """ Remove issues with b' or b" prepended to strings with unicode encoding
        run with Python 3
    """
    new_value = value
    if isinstance(value, str):
        if value.startswith("b'") or value.startswith('b"'):
            new_value = value[2:-1]
    return new_value
```

```python
# Slim down metadata
import pandas as pd

fpath = '/projects/AO3Scraper/output/ao3_dcu_text/stories.csv'
metadata = pd.read_csv(fpath)
```

```python
metadata = metadata.applymap(remove_bytestring)
metadata
```

```python
metadata['fic_id'] = pd.to_numeric(metadata.fic_id)
```

```python
metadata.drop_duplicates(subset='fic_id', inplace=True)
metadata.shape
```

```python
filtered = metadata[metadata['fic_id'].isin(story_ids)]
filtered
```

```python
# Save out
filtered.to_csv(fpath, index=False)
```

## Already scraped is in complete_en_1k-50k stories
Now in preprocess_fics.py

```python
# Load fic IDs of data already scraped
import pandas as pd

fandom = 'dcu'
scraped_metadata_path = f'/data/fanfiction_ao3/{fandom}/complete_en_1k-50k/metadata.csv'
existing_metadata = pd.read_csv(scraped_metadata_path)
existing_metadata
```

```python
# Load scraped fic IDs
import os
import pandas as pd

scraped_data_dirpath = '/projects/AO3Scraper/output/'
all_fic_ids = pd.DataFrame()
for segment in ['teen', 'general', 'explicit', 'mature', 'unrated']:
    print(segment)
    fic_ids_path = os.path.join(scraped_data_dirpath, f'{fandom}_fic_ids_{segment}.csv')
    fic_ids_data = pd.read_csv(fic_ids_path, index_col=0, header=None)
    print(len(fic_ids_data))
    new_fic_ids = list(set(fic_ids_data.index) - set(existing_metadata['fic_id']))
    print(len(new_fic_ids))
    new_data = fic_ids_data.loc[new_fic_ids,:]
    all_fic_ids = pd.concat([all_fic_ids, new_data])
    print()
    
# Save out new fic ids
outpath = os.path.join(scraped_data_dirpath, f'{fandom}_new_fic_ids.csv')
print(len(all_fic_ids))
all_fic_ids.to_csv(outpath, header=False)
```
