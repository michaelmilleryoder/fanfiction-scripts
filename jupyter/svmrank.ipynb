## Remove e lines

```python
output_fpath = '/projects/fanfiction-nlp/quote_attribution/tmp/pride_prejudice.csv/svmrank_input_modified.txt'

outlines = []
for group in groups:
    for line in group:
        newline = []
        for val in line:
            if 'e' in val:
                idx = val.split(':')[0]
                new_val = f'{idx}:0.0'
                newline.append(new_val)
            else:
                newline.append(val)
        #newline = [0.0 if 'e' in val else val for val in line]
        outlines.append(newline)
        
with open(output_fpath, 'w') as f:
    for line in outlines:
        f.write('\t'.join(line) + '\n')
```

# Inspect svmrank_input.txt

```python
bad_fpath = '/projects/fanfiction-nlp/quote_attribution/tmp/pride_prejudice.csv/svmrank_input.txt'
groups = []

with open(bad_fpath, 'r') as f:
    lines = f.read().splitlines()
    group_lines = []
    for line in lines:
        if line.startswith('#'):
           # Add lines to the group 
            if len(group_lines) == 0:
                continue
            groups.append(group_lines)
            group_lines = []
        else:
            group_lines.append(line.split('\t'))
            
len(groups)
```

```python
# Check for answers in each group
for group in groups:
    answers = [l[0] for l in group]
    if not '1' in answers:
        print("no match")
```
