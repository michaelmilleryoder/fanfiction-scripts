# Filter to fics of manageable size, create options

```python
# Filter to dataset fics of manageable size
import os
import pandas as pd

fandoms = [
    'allmarvel',
    'supernatural',
    'harrypotter',
    'dcu',
    'sherlock',
    'teenwolf',
    'starwars',
    'drwho',
    'tolkien',
    'dragonage',
]

metadata = {}

# Build combined metadata
for fandom in fandoms:
    metadata_fpath = os.path.join('/data/fanfiction_ao3', fandom, 'complete_en_1k-50k', 'metadata.csv')
    metadata[fandom] = pd.read_csv(metadata_fpath)

all_metadata = pd.concat(metadata.values())
all_metadata
```

```python
filtered = all_metadata[(all_metadata['words']<5000) & (all_metadata['rating'] != '["Explicit"]')]
filtered
```

# Find fics with explicit and implicit queer identity labels

```python
import re

queer_identity_labels = [
    r'\bqueer\b',
    r'\bgay\b',
    r'\blesbian\b',
    r'\baro\b',
    r'\bace\b',
    r'\btrans\b',
    r'\bbi\b',
    r'\bnonbinary\b',
    r'\bnon-binary\b'
]

queer_relationship_types = [
    'M/M',
    'F/F'
]

explicit_filter = filtered['additional tags'].map(lambda x: any([re.search(w, x.lower()) for w in queer_identity_labels]))
explicit = filtered[explicit_filter]
print(len(explicit))

implicit_filter = filtered['category'].map(lambda x: any([t in x for t in queer_relationship_types])) & [not val for val in explicit_filter.tolist()]
implicit = filtered[implicit_filter]
print(len(implicit))
```

```python
# Sample from explicit
pd.set_option('display.max_colwidth', None)
explicit.sample(20).loc[:, ['fic_id', 'title', 'rating', 'category', 'fandom', 'additional tags', 'character', 'relationship', 'words']]
```

```python
# Sample from implicit
implicit.sample(20).loc[:, ['fic_id', 'title', 'rating', 'category', 'fandom', 'additional tags', 'character', 'relationship', 'words']]
```

```python
implicit[implicit['fandom'].map(lambda x: 'Sherlock' in x) & (implicit['additional tags'].map(lambda x: 'domestic' in x))].sample(10)
```

# Find fics with specific qualities

```python
# Find fluff F/M fic
selected = filtered[(filtered['category'].map(lambda x: 'F/M' in x)) & (filtered['additional tags'].map(lambda x: 'Fluff' in x))]
selected.sample(10)
```

```python
# Find Tolkien F/F fic
selected = filtered[(filtered['category'].map(lambda x: 'F/F' in x)) & (filtered['fandom'].map(lambda x: 'tolkien' in x.lower()))]
selected.sample(10)
```
