# Load fanfiction tags data

```python
# path = '/projects/fanfiction_lgbtq/output/fanfic_label_proportions.csv' # from fanfiction_lgbtq_tags_over_time.ipynb
# fanfic_props = pd.read_csv(path, index_col=['published'], parse_dates=True)
path = '/projects/fanfiction_lgbtq/output/fanfic_label_proportions_datasets.csv' # from fanfiction_lgbtq_tags_over_time.ipynb
fanfic_props = pd.read_csv(path, index_col=['dataset','published'], parse_dates=True)
# print(fanfic_props.columns)
# print(fanfic_props.index)
# fanfic_props

# Transform to wide format for dataset split
fanfic_props = fanfic_props.pivot_table(index=['published'], columns=['dataset'])
fanfic_props.columns = ['_'.join(col) for col in fanfic_props.columns.values]
# fanfic_props

## Fit b-spline curves to fanfiction tags
import numpy as np
import scipy.interpolate as interpolate
import plotly.express as px

propcols = [col for col in fanfic_props if 'proportion' in col]
fanfic_props['rownum'] = range(len(fanfic_props))
default_smooth = 0.0001
smooth = { # smoothing parameters
   'proportion_hurt/comfort': 0.0005,
   'proportion_trans': 0.0002,
   'proportion_wedding-marriage_M/M-F/F': 0.0001,
}
# smooth = { # smoothing parameters
#    'proportion_hurt/comfort_lo_lgbtq': 0.0005,
#    'proportion_hurt/comfort_hi_lgbtq': 0.0005,
#    'proportion_trans_lo_lgbtq': 0.0002,
#    'proportion_trans_hi_lgbtq': 0.0002,
#    'proportion_wedding-marriage_M/M-F/F_lo_lgbtq': 0.0001,
#    'proportion_wedding-marriage_M/M-F/F_hi_lgbtq': 0.0001,
# }
for col in propcols:
#     t, c, k = interpolate.splrep(fanfic_props['rownum'], fanfic_props[col], s=0.0001, k=3)
    t, c, k = interpolate.splrep(fanfic_props['rownum'], fanfic_props[col], s=smooth.get(col, default_smooth), k=3)
    N = 44 # number of fake data points
    xx = np.linspace(fanfic_props.rownum.min(), fanfic_props.rownum.max(), N)
    spline = interpolate.BSpline(t,c,k, extrapolate=False)
#     if col == 'proportion_hurt/comfort':
    if 'proportion_trans' in col:
        print(col)
#     if col == 'proportion_wedding-marriage_M/M-F/F':
        fig = px.line(y=[fanfic_props[col], spline(xx)])
        fig.show()
    fanfic_props[col] = spline(xx)
```

# Shift curves, measure correlations

```python
# Shift fanfiction tag probabilities
from IPython.display import display
# fanfic_cols = [col for col in merged.columns if 'proportion' in col]
# news_cols = [col for col in merged.columns if 'proportion' not in col and col !='date']
comparisons = {
    'proportion_trans': ['trans/queer_gender_issues', 'american_politics'],
    'proportion_wedding-marriage_M/M-F/F': ['law/same-sex_marriage', 'american_politics']
}
titles = {
    'proportion_trans': 'Fanfiction trans tag correlation, shifted',
    'proportion_wedding-marriage_M/M-F/F': 'Fanfiction gay marriage tag correlation, shifted',
}
save_names = {
    'proportion_trans': 'trans',
    'proportion_wedding-marriage_M/M-F/F': 'gay_marriage',
}

for col in ['proportion_trans', 'proportion_wedding-marriage_M/M-F/F']:
    selected_fanfic = []
    for shift in range(-6, 7):
#         colname = f'{col}_shift{shift*3}'
        colname = shift*3
        merged[colname] = merged[col].shift(shift)
        selected_fanfic.append(colname)
    selected_news = comparisons[col]
    corr = merged[selected_fanfic + selected_news].corr()
    display(corr.loc[selected_news, selected_fanfic])
    merged.drop(columns=colname, inplace=True)

    # Plot
    import plotly.express as px
    fig = px.line(corr.loc[selected_fanfic, selected_news], y=selected_news, title=titles[col], 
                 labels=dict(index='month shift', value='correlation'))
    fig.update_xaxes(tickvals=list(range(-18,19,6)))
    fig.update_layout(
        legend=dict(yanchor="top", y=1.3, xanchor='left', x=0.6)
    )
    fig.add_vline(x=0)
    fig.write_image(f'/projects/fanfiction_lgbtq/output/plots/{save_names[col]}_shifted.pdf')
    fig.show()
```

## Residual stuff
do a regression from average number of fanfiction tags to fanfiction trans tag prevalence and then take the residual.  This will tell you what prevalence of that tag is not explained by the general prevalence of tags.  And then use the residual to predict the trans news topic prevalence

```python
import statsmodels.api as sm

# fanfic_col = 'proportion_trans'
# fanfic_col = 'proportion_trans_lo_lgbtq'
# fanfic_col = 'proportion_wedding-marriage_M/M-F/F'
fanfic_col = 'proportion_wedding-marriage_M/M-F/F_hi_lgbtq'
# fanfic_col = 'proportion_wedding-marriage_M/M-F/F_lo_lgbtq'
# news_col = 'american_politics'
news_col = 'pride/black_identity'

# X = sm.add_constant(merged[['n_tags_lo_lgbtq']])
X = sm.add_constant(merged[['n_tags_hi_lgbtq']])
mod = sm.OLS(merged[fanfic_col], X)
res = mod.fit()
resid = res.resid

mod = sm.OLS(merged[news_col], sm.add_constant(resid))
res = mod.fit()
res.summary()
```
