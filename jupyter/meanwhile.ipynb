```python
# Combine article texts
from tqdm.notebook import tqdm
import os
import pandas as pd

parts = []
out_dirpath = '/data/news/now2010-2021/lgbtq/text/'
for fname in tqdm(sorted(os.listdir(out_dirpath))[:100]):
    fpath = os.path.join(out_dirpath, fname)
    parts.append(pd.read_csv(fpath))
all_parts = pd.concat(parts).set_index('article_id')
print(all_parts.shape)
all_parts.head()
```

```python
# Preprocess text
pd.set_option('display.max_colwidth', None)
all_parts['text'] = all_parts.text.str.replace(r'<.*?>', '')
all_parts['text'] = all_parts.text.str.replace(r'@ ', '')
all_parts.head()
```

```python
# Add in article metadata

# Load article metadata
sources = pd.read_csv('/data/news/now2010-2021/lgbtq/sources.csv', index_col=0, parse_dates=['date'])
print(sources.shape)
sources.head()
```

```python
merged = pd.merge(sources, all_parts, left_index=True, right_index=True)
print(merged.shape)
merged.head()
```

```python
# Save out
merged.to_csv('/data/news/now2010-2021/lgbtq/articles100.csv')
```
