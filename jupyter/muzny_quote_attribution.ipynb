## Build token file

```python
# Lemmatize, postag
import re
import spacy
from spacy.tokenizer import Tokenizer

nlp = spacy.load('en')
nlp.tokenizer = Tokenizer(nlp.vocab, token_match=re.compile(r'\S+').match)
fic_data['token'] = fic_data.text_tokenized.map(nlp)
fic_data['token']
# processed = nlp(fic_data.loc[0, 'text_tokenized'])
# processed

fic_data['postag'] = [[tok.tag_ for tok in toks] for toks in fic_data['token'].tolist()]
fic_data['lemma'] = [[tok.lemma_ for tok in toks] for toks in fic_data['token'].tolist()]

def get_sent_ids(toks):
    sents = list(toks.sents)
    return [sents.index(tok.sent) + 1 for tok in toks]
    
fic_data['sent_id'] = fic_data['token'].map(get_sent_ids)

# Need to get the -1 for roots
fic_data['head_id'] = [[tok.head.i for tok in toks] for toks in fic_data['token'].tolist()]
fic_data['deprel'] = [[tok.dep_ for tok in toks] for toks in fic_data['token'].tolist()]
fic_data['token_id'] = [[tok.i + 1 for tok in toks] for toks in fic_data['token'].tolist()]

# melt on whitespace for string columns
explode_cols = ['token', 'postag', 'lemma', 'sent_id', 'head_id', 'deprel', 'token_id']
toks = fic_data.apply(lambda x: x.explode() if x.name in explode_cols else x)

# Renumber ID columns to continue throughout the df instead of repeating
toks['fic_sent_id'] = toks.groupby(['para_id', 'sent_id']).ngroup()
toks['fic_token_id'] = range(len(toks))
toks

# Rename columns
transform = {
    'para_id': 'paragraphId', 
    'fic_sent_id': 'sentenceID', 
    'fic_token_id': 'tokenId', 
    'head_id': 'headTokenId',
    'token': 'originalWord',
    'postag': 'pos',
}

tok_data = toks.rename(columns=transform)
tok_data = tok_data[[col for col in tok_input.columns if col in tok_data.columns]]
tok_data
# toks
```

## Build coref input

```python
# Load annotated fanfiction coref output
import pdb

class AnnotatedSpan():
    def __init__(self,
            chap_id=None,
            para_id=None,
            start_token_id=None,
            end_token_id=None,
            annotation=None, # speaker for quotes, or character for character mentions
            text=''):
        self.chap_id = chap_id # starts with 1, just like annotations
        self.para_id = para_id # starts with 1, just like annotations
        self.start_token_id = start_token_id # starts over every paragraph, starts with 1 just like annotations
        self.end_token_id = end_token_id
        self.annotation = annotation
        self.text = text
        self.text_tokens = None
        
    def __repr__(self):
        return f"{self.chap_id}.{self.para_id}.{self.start_token_id}-{self.end_token_id},annotation={self.annotation}"


        
annotations = []
annotations_set = set()
coref_dirpath = '/data/fanfiction_ao3/annotated_10fandom/test/entity_clusters/'
for fname in sorted(os.listdir(coref_dirpath))[:1]:
    df = pd.read_csv(os.path.join(coref_dirpath, fname))
    for colname in df.columns: # each column is a character
            annotations_set.add(colname)
            for mention in df[colname].dropna():
                parts = mention.split('.')
                chapter_id = int(parts[0])
                paragraph_id = int(parts[1])
                if '-' in parts[2]:
                    token_id_start = int(parts[2].split('-')[0])
                    token_id_end = int(parts[2].split('-')[-1])
                else:
                    token_id_start = int(parts[2])
                    token_id_end = int(parts[2])
                annotations.append(AnnotatedSpan(
                  chap_id=chapter_id,
                  para_id=paragraph_id,
                  start_token_id=token_id_start,
                  end_token_id=token_id_end,
                  annotation=colname
              ))
                    
print(len(annotations))
print(len(annotations_set))
```

```python
# Build mapping from tokenId to coreference_ids
coref_ents['span'] = [range(beg, end+1) for (beg, end) in zip(coref_ents.span_start, coref_ents.span_end)]
coref_ents_toks = coref_ents.explode('span')
tok2char = coref_ents_toks.set_index('span')['coreference_id'].to_dict()
tok_data['characterId'] = tok_data.tokenId.map(lambda x: tok2char.get(x, -1))
tok_data
```

## Add quote information to token input file

```python
# Load quotes
quotes = []
quotes_set = set()
coref_dirpath = '/data/fanfiction_ao3/annotated_10fandom/test/quote_attribution/'
for fname in sorted(os.listdir(coref_dirpath))[:3]:
    if 'qinlan' not in fname and 'michael' not in fname:
        print(fname)
        df = pd.read_csv(os.path.join(coref_dirpath, fname))
        for colname in df.columns: # each column is a character
                quotes_set.add(colname)
                for mention in df[colname].dropna():
                    parts = mention.split('.')
                    chapter_id = int(parts[0])
                    paragraph_id = int(parts[1])
                    if '-' in parts[2]:
                        token_id_start = int(parts[2].split('-')[0])
                        token_id_end = int(parts[2].split('-')[-1])
                    else:
                        token_id_start = int(parts[2])
                        token_id_end = int(parts[2])
                    quotes.append(AnnotatedSpan(
                      chap_id=chapter_id,
                      para_id=paragraph_id,
                      start_token_id=token_id_start,
                      end_token_id=token_id_end,
                      annotation=colname
                  ))
print(len(quotes))
print(len(quotes_set))
```

```python
# Turn annotations into df
quote_data = pd.DataFrame([[annotation.annotation, annotation.text, annotation.chap_id, annotation.para_id, annotation.start_token_id, annotation.end_token_id] \
                           for annotation in quotes],
                         columns=['character_name', 'tokens', 'chap_id', 'para_id', 'start_token_id', 'end_token_id'])
quote_data['span_start'] = [global_token_id[el] for el in tuple(zip(quote_data.chap_id, quote_data.para_id, quote_data.start_token_id))]
quote_data['span_end'] = [global_token_id[el] for el in tuple(zip(quote_data.chap_id, quote_data.para_id, quote_data.end_token_id))]

# Build mapping from tokenId to coreference_ids
quote_data['span'] = [range(beg, end+1) for (beg, end) in zip(quote_data.span_start, quote_data.span_end)]

quote_data_toks = quote_data.explode('span')
quote_data_toks

quote_data_toks.reset_index(inplace=True)
quote_data_toks.rename(columns={'index': 'quote_id'}, inplace=True)
quote_data_toks

beg_quotes = quote_data_toks.groupby('quote_id').head(1).index
quote_data_toks.loc[beg_quotes, 'inQuotation'] = 'B-QUOTE'

quote_data_toks['inQuotation'] = quote_data_toks.inQuotation.fillna('I-QUOTE')

tok2quote = quote_data_toks.set_index('span')['inQuotation'].to_dict()

tok_data['inQuotation'] = tok_data.tokenId.map(lambda x: tok2quote.get(x, 'O'))
tok_data
```
